```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leer el archivo CSV en un DataFrame
df = pd.read_csv('fake-news/train.csv')

# Mostrar las primeras filas del DataFrame
print(df.head())

# Resumen estadístico de las columnas numéricas (en este caso solo 'label')
print(df.describe())

# Distribución de las etiquetas (noticias falsas/verdaderas)
plt.figure(figsize=(6, 4))
sns.countplot(x='label', data=df)
plt.title('Distribución de Etiquetas')
plt.xlabel('Etiqueta')
plt.ylabel('Cantidad')
plt.show()

# Calcular longitud de los títulos y textos

df['Length_Title'] = df['title'].astype(str).apply(len)
df['Length_Text'] = df['text'].astype(str).apply(len)

# Diagrama de dispersión: Longitud del título vs. Longitud del texto
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Length_Title', y='Length_Text', hue='label', data=df)
plt.title('Longitud del Texto vs. Longitud del Título')
plt.xlabel('Longitud del Título')
plt.ylabel('Longitud del Texto')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Importamos los datos 
df = pd.read_csv('fake-news/train.csv')

# Mostrar las primeras filas del DataFrame
print(df.head())

# Convertir la columna 'text' a tipo cadena (string) y luego calcular la longitud
df['text_length'] = df['text'].astype(str).apply(len)

# Visualizar la distribución de la longitud del texto mediante un histograma
plt.figure(figsize=(8, 6))
plt.hist(df['text_length'], bins=20, color='skyblue')
plt.xlabel('Longitud del Texto')
plt.ylabel('Frecuencia')
plt.title('Histograma de Longitud del Texto de los Artículos')
plt.show()
```

```python
from scipy import stats

# Mostrar histograma de la longitud de los textos
plt.figure(figsize=(8, 6))
plt.hist(df['text_length'], bins=30, alpha=0.7)
plt.xlabel('Longitud del Texto')
plt.ylabel('Frecuencia')
plt.title('Distribución de Longitud de Textos')
plt.show()

# Prueba de normalidad (Shapiro-Wilk)
stat, p = stats.shapiro(df['text_length'])
print(f'Estadístico de prueba: {stat}, Valor p: {p}')

alpha = 0.05
if p > alpha:
    print('Los datos de longitud de texto parecen seguir una distribución normal (no se rechaza H0)')
else:
    print('Los datos de longitud de texto no siguen una distribución normal (se rechaza H0)')
```
